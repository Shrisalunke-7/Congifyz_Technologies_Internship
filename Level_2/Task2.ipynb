```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://drive.google.com/file/d/1dKZxapT3xLzOTOpy1LCyqctL8YEvzp4Y/view?usp=drive_link"
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_csv(url)
df.head()
```

# Task2: Cuisine Combination

**Q.1) Identify the most common combinations of
cuisines in the dataset.**

```python
# Count the occurrences of each combination of cuisines
cuisine_combinations = df['Cuisines'].value_counts()

# Display the most common combinations
print("Most Common Combinations of Cuisines:")
print(cuisine_combinations)
```

```python
top_20_commons=cuisine_combinations.head(20)
print(f"Top 20 Most Common Combinations of Cuisines : {top_20_commons}")
```

**Q.2) Determine if certain cuisine combinations
tend to have higher ratings.**

```python
# Calculate the average rating for each cuisine combination
average_ratings = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
average_ratings
```

```python
# Display the average ratings for each cuisine combination
print(f"The Higher Rating by Cuisine Combinations is: {average_ratings.max()}")
```
