```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://drive.google.com/file/d/1dKZxapT3xLzOTOpy1LCyqctL8YEvzp4Y/view?usp=drive_link"
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_csv(url)
df.head()
```

# Task3: Price Range Distribution

**Q.1) Create a histogram or bar chart to
visualize the distribution of price ranges
among the restaurants.**

```python
# Bar Chart

# Count the occurrences of each price range
price_range_counts = df['Price range'].value_counts()

# Plot the distribution of price ranges using a bar chart
price_range_counts.plot(kind='bar', color='blue')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.show()
```

**Q.2) Calculate the percentage of restaurants
in each price range category.**

```python
# Count the occurrences of each price range
price_range_counts = df['Price range'].value_counts()

# Calculate the percentage of restaurants in each price range
price_range_percentages =round( (price_range_counts / len(df)) * 100)

# To Display the percentage of restaurants in each price range
print("Percentage of Restaurants in Each Price Range:")
print(price_range_percentages)
```

```python
# Plot the distribution of price ranges using a pie chart
plt.figure(figsize=(8, 5))
plt.pie(price_range_percentages, labels=price_range_percentages.index, autopct='%1.1f%%', colors=['red', 'green', 'blue', 'skyblue'])
plt.title('Percentage of Restaurants in Each Price Range')
plt.show()
```
